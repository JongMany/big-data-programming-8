```python
import pandas as pd

# 파일 경로
file_paths = {
    "전철역 개수": "results/지하철 역 개수.csv",
    "elevators_wheelchair_lifts": "results/elevators_wheelchair_lifts.csv",
    "구별_충전소개수": "results/구별_충전소개수.csv",
    "구별_평균연단간격": "results/구별_평균연단간격.csv",
    "대기시간": "results/대기시간.csv",
    "의료시설점수": "results/의료시설점수.csv",
    "체육시설점수": "results/체육시설점수.csv",
    "인구수": "results/인구수.csv",
    "저상버스점수": "results/저상버스비율2.csv"
}

# 기본 데이터프레임 설정
base_df = pd.read_csv(file_paths["전철역 개수"])

# 나머지 파일을 읽어와서 병합
for key, file_path in file_paths.items():
    if key != "전철역 개수":
        temp_df = pd.read_csv(file_path)
        
        # '구' 열이 있는지 확인
        if '구' in temp_df.columns:
            base_df = pd.merge(base_df, temp_df, on='구', how='left')
        else:
            print(f"파일 {file_path} 에 '구' 열이 없습니다.")

# 병합되지 않은 구에 대해 처리
missing_data_df = base_df[base_df.isnull().any(axis=1)]
print("병합되지 않은 데이터:")
print(missing_data_df)

# # 병합된 결과를 새로운 CSV 파일로 저장
base_df.to_csv("results/merged_data.csv", index=False, encoding='utf-8')

base_df
```

```python
# 데이터 프레임의 열 순서를 변경하고, 특정 열을 삭제합니다.
base_df = base_df[['구', '지체장애인인구수', '저상버스비율2', '콜택시대기초', '연단간격점수평균', '엘리베이터개수평균', '휠체어리프트개수평균', '충전소개수', '의료시설점수', '체육시설점수', 
'전철역 개수']]
asd = base_df

asd = asd.set_index('구')

asd = asd.astype(float)

asd['충전소개수/지체장애인인구수'] = asd['충전소개수'] / asd['지체장애인인구수']

asd
```

```python
import pandas as pd

# 상관관계 행렬 계산
correlation_matrix = asd.corr()

# 상관관계 행렬 출력
print(correlation_matrix)

# 상관관계 행렬 시각화
import seaborn as sns
import matplotlib.pyplot as plt


plt.rcParams['font.family'] ='AppleGothic'
plt.rcParams['axes.unicode_minus'] =False

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
sdf = asd

import pandas as pd
from sklearn.preprocessing import StandardScaler

# 정규화 수행
scaler = StandardScaler()
df_scaled = scaler.fit_transform(sdf)

# 정규화된 데이터프레임 생성
df_scaled = pd.DataFrame(df_scaled, columns=sdf.columns, index=sdf.index)

# 결과 출력
df_scaled


```

```python
연단 = 0.2
엘베 = 0.7
휠체어리프트 = 0.1

df_scaled['지하철 이동 지수'] = df_scaled['연단간격점수평균'] * 연단 + df_scaled['엘리베이터개수평균'] * 엘베 + df_scaled['휠체어리프트개수평균'] * 휠체어리프트

전휠충전 = 0.2
체육시설 = 0.3
의료시설 = 0.5

df_scaled['편의시설 점수'] = df_scaled['충전소개수'] * 전휠충전 + df_scaled['체육시설점수'] * 체육시설 + df_scaled['의료시설점수'] * 의료시설

df_scaled
```

```python
df2 = df_scaled[['저상버스비율2', '콜택시대기초', '지하철 이동 지수', '편의시설 점수']].copy()

# df2.rename(columns={'콜택시대기시간(초단위)': '콜택시대기시간 점수'}, inplace=True)

# 정규화 수행
df_scaled2 = scaler.fit_transform(df2)

# 정규화된 데이터프레임 생성
df_scaled2 = pd.DataFrame(df_scaled2, columns=df2.columns, index=df2.index)

print(len(df_scaled2))

df_scaled2
```

```python
# 종민님 자료 근거 (무슨 자료지?)
a = 26.5
b = 7.2
c = 18.2
sum = a + b + c

저상버스비율2 = a/sum
콜택시대기시간 = b/sum
지하철 = c/sum

df_scaled3 = df_scaled2

df_scaled3['대중교통 이동성 점수'] = df_scaled2['저상버스비율2'] * 저상버스비율2 + df_scaled2['콜택시대기초'] * 콜택시대기시간 + df_scaled2['지하철 이동 지수'] * 지하철

df_scaled3
```

```python
# 정규화 수행
df_scaled4 = scaler.fit_transform(df_scaled3)

# 정규화된 데이터프레임 생성
df_scaled5 = pd.DataFrame(df_scaled4, columns=df_scaled3.columns, index=df_scaled3.index)

df_scaled5
```

```python
편의시설 = 0.3
대중교통이동성점수 = 0.7

df_scaled5['최종 점수'] = df_scaled5['편의시설 점수'] * 편의시설 + df_scaled5['대중교통 이동성 점수'] * 대중교통이동성점수

df_scaled6 = scaler.fit_transform(df_scaled5)

df_scaled6 = pd.DataFrame(df_scaled6, columns=df_scaled5.columns, index=df_scaled5.index)

df_scaled6
```

```python
df_scaled7 = df_scaled6[['최종 점수']]

print(len(df_scaled7))

df_scaled8 = df_scaled7.sort_values(by='최종 점수', ascending=False)

df_scaled8
```
