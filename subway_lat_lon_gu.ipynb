```python
import re

df2 = df.copy()

# 정규표현식으로 전철역명 변경
# '전철역명' 열의 문자열 끝에 '역'이 있는 경우 이를 제거
df2['전철역명'] = df2['전철역명'].apply(lambda x: re.sub('역$', '', x))

df2['전철역명'] = df2['전철역명'].apply(lambda x : re.sub('[?]','.',x))

df2['호선'] = df2['호선'].apply(lambda x: re.sub('^0', '', x))
# data['호선'] = data['호선'].apply(lambda x : x.replace('0',''))

df2
```

```python
search_list = df2['검색어'].tolist()
search_list
```

```python
import requests
from urllib.parse import urlparse

# 주소 검색 함수 정의
def address_search(address):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json?&query=" + address + "&category_group_code=SW8"
    result = requests.get(urlparse(url).geturl(),
                          headers={"Authorization":"KakaoAK 5b16219c37dbd9eb4fc13280d64b16dd"})
    json_obj = result.json()
    global val   # 전역 변수를 지역 범위에서 사용하고 싶을 경우
    for document in json_obj['documents']:
        # 값이 NoneType 일 경우의 오류 방지
        try:
            val = [document['road_address_name']]
        except:
            val = [0]
    return val
```

```python
address_search('서울역 공항철도')   # 서울역 경의선 제외 모두 이상 -> 추후 변경 예정
```

```python
# 주소 검색 함수 반복 적용
list1 = []
for i in search_list:
    list1.append(address_search(i))
```

```python
data2 = pd.DataFrame(list1, columns=['주소'])
data2.head()
```

```python
data3 = pd.concat([df2, data2], axis=1)
print(data3.shape)
data3.head()
```
